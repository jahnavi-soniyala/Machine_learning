```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df= pd.read_csv("lrdata.csv")
df
```

```python
from sklearn.model_selection import train_test_split
x=df["year"]
y=df["per capita income (US$)"]
#dfx_train,dfx_test,dfy_train,dfy_test=train_test_split(x,y,test_size=0.2,random_state=42)
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
xtr=pd.DataFrame(X_train)
xte=pd.DataFrame(X_test)
ytr=pd.DataFrame(y_train)
yte=pd.DataFrame(y_test)
```

```python
model = linear_model.LinearRegression()
model.fit(xtr, ytr)
```

```python
df_predict=pd.DataFrame(model.predict(xte))
df_predict
```

```python
mse = mean_squared_error(yte,df_predict[0])
r2 = r2_score(yte,df_predict[0])

print("Mean Squared Error:", int(mse))
print("R^2 Score:", int(r2))
```

```python
plt.scatter(df['year'],df['per capita income (US$)'],color='red',marker="*")
plt.plot(X_test,df_predict[0])
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error 
from sklearn import preprocessing

# Importing data
df = pd.read_csv('Real_estate.csv') 
sns.scatterplot(x='X4 number of convenience stores', y='Y house price of unit area', data=df)
plt.title('Scatterplot of Convenience Stores vs House Price')
plt.xlabel('Number of Convenience Stores')
plt.ylabel('House Price of Unit Area')
plt.show()

# Creating feature variables and target variable
X = df.drop('Y house price of unit area', axis=1) 
y = df['Y house price of unit area']

# Display the features and target variable
print("Features:\n", X.head()) 
print("Target Variable:\n", y.head())

# Creating train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

# Creating a regression model 
model = LinearRegression()

# Fitting the model 
model.fit(X_train, y_train)

# Making predictions
predictions = model.predict(X_test)

# Model evaluation
mse = mean_squared_error(y_test, predictions)
mae = mean_absolute_error(y_test, predictions)
print('Mean Squared Error:', mse) 
print('Mean Absolute Error:', mae)

# Optional: Plotting predicted vs actual values
plt.figure(figsize=(10, 6))
plt.scatter(y_test, predictions)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual vs Predicted House Prices')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], '--r')  # Line for perfect predictions
plt.show()
```
